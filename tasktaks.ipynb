```python
plt.figure(figsize=(15,15))
plt.rcParams['figure.figsize']=(12.8, 7.2)

# 
origin_img = cv.imread('task.png', 0)
img = cv.resize(origin_img, (0, 0), None, fx=0.3, fy=0.3)
img = cv.medianBlur(img, 5)

# 
circles = cv.HoughCircles(img, cv.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=95, maxRadius=120)

# 
new_window = np.zeros_like(img)

# 
mask = np.zeros_like(img)

if circles is not None:
    circles = np.uint16(np.around(circles))
    
    for i in circles[0, :]:
        center = (i[0], i[1])
        radius = i[2]

        # 
        cv.circle(new_window, center, radius, (255, 255, 255), thickness=cv.FILLED)

        cv.circle(mask, center, radius, (255, 255, 255), thickness=cv.FILLED)

        # 
        contour_image = cv.bitwise_and(img, img, mask=mask)
        
plt.subplot(3,2,1)
plt.xticks([]),plt.yticks([])
plt.imshow(contour_image,cmap='gray')

kernel = np.ones((3,3),np.uint8)
erosion = cv.erode(contour_image,kernel,iterations = 6)

plt.subplot(3,2,2)
plt.xticks([]),plt.yticks([])
plt.imshow(erosion,cmap='gray')
```

```python
plt.figure(figsize=(15,15))
plt.rcParams['figure.figsize']=(12.8, 7.2)

img = cv.resize(erosion,(0,0),None,fx=0.3,fy=0.3)
img = cv.medianBlur(img,5)
cimg = cv.cvtColor(img,cv.COLOR_BGR2RGB)
circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,param1=50,param2=30,minRadius=20,maxRadius=40)

if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # 
        cv.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 2)
        cv.circle(img, (i[0], i[1]), i[2], (255, 0, 0), 2)
        cv.circle(origin_img,(i[0], i[1]), i[2], (0, 255, 0), 2)
        # 
        cv.circle(cimg, (i[0], i[1]), 2, (0, 0, 255), 3)

plt.subplot(3,3,1)
plt.title("Color")
plt.xticks([]),plt.yticks([])
plt.imshow(cimg)

plt.subplot(3,3,2)
plt.title(" ")
plt.xticks([]),plt.yticks([])
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)

plt.subplot(3,3,3)
plt.title(" ")
plt.xticks([]),plt.yticks([])
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(origin_img)
```

```python
ret,thresh1 = cv.threshold(cimg,150,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(cimg,150,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(cimg,150,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(cimg,150,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(cimg,120,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [cimg, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
ret,thresh1 = cv.threshold(cimg,150,255,0)
# 
contours, _ = cv.findContours(thresh1, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

area_threshold = 40 # 

for contour in contours:
    area = cv.contourArea(contour)
    if area > area_threshold:
        # 
        M = cv.moments(contour)
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])

        # 
        text = "Volleyball"
        cv.putText(cimg, text, (cX, cY), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 3)
```
